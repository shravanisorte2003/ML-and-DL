```python
#loading the library with the iris dataset
from sklearn.datasets import load_iris
```

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
np.random.seed(0)
```

```python
#creating an object called iris with the iris data
iris=load_iris()
#creating a dataframe with the four features variables
df=pd.DataFrame(iris.data,columns=iris.feature_names)
#viewing the top 5 rows
df.head()
```

```python
print(iris)
```

```python
#adding a new column for the species name
df['species']=pd.Categorical.from_codes(iris.target,iris.target_names)
df.head()
```

```python
#creating test and train data
df['is_train']=np.random.uniform(0,1,len(df))<=.75
df.head()
```

```python
#creating dataframe with test rows and training rows
train,test=df[df['is_train']==True],df[df['is_train']==False]
print('no. of observations in the training data: ',len(train))
print('no. of observation in the test adat: ',len(test))
```

```python
#create a list of the feature column's names
features=df.columns[:4]
features
```

```python
#converting each species name into digits
y=pd.factorize(train['species'])[0]
#viewing target
y
#0,1,2 represents 3 different kinds of flower we have
```

```python
#creating a random forest classifier
clf=RandomForestClassifier(n_jobs=2,random_state=0)
#training the classifier
clf.fit(train[features],y)
```

```python
features
```

```python
test[features]
```

```python
#Applying the training classifier to the test
clf.predict(test[features])
```

```python
#viewing the predicted probabilities of the first 10 observations
clf.predict_proba(test[features])[0:10]
```

```python
#mapping names for the plants for each predicted plant class
preds=iris.target_names[clf.predict(test[features])]
preds[0:51]
```

```python
#viewing the actual species for the first five observations
test['species'].head(41)
```

```python
#creating confusion matrix
pd.crosstab(test['species'],preds,rownames=['Actual Species'],colnames=['Predicted Species'])
```
